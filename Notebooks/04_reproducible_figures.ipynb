# Reproducible figures and the GUI
This notebook is a lightweight companion to the GUI.

The GUI is for exploration; this notebook shows how to reproduce a figure programmatically so you can use it in papers or notes.

````python
from _bootstrap import add_src_to_path
repo_root = add_src_to_path()

import numpy as np
import matplotlib.pyplot as plt

from rayleigh_gans.form_factors import (
    q_from_theta,
    F_sphere_q,
)

plt.rcParams["figure.dpi"] = 120
````

## Example: save a publication-style PNG from Python
This mirrors what the GUI "Save plot" button does, but in a scripted way.

````python
R = 250.0
nm = 1.33
wavelength = 532.0
theta = np.linspace(0, np.pi, 1800)
q = q_from_theta(theta, nm=nm, wavelength=wavelength)

F = F_sphere_q(q, R)
I = np.abs(F)**2

plt.figure(figsize=(7.2,4.8))
plt.plot(q, I, linewidth=2.2)
plt.xlabel(r"$q$ [1/nm]")
plt.ylabel(r"$|F(q)|^2$")
plt.title(f"Sphere intensity (R={R:g} nm, Î»={wavelength:g} nm, n={nm:g})")
plt.grid(True, alpha=0.25)

out = "sphere_intensity_example.png"
plt.savefig(out, dpi=300, bbox_inches="tight")
out
````
